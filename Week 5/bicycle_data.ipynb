This notebook is used to exercise inclass solutions for the NL bicycle dataset

```python
import pandas as pd

# Load the Excel file
df = pd.read_excel('NL_bicycle_sales.xlsx')

# Display the first few rows
print("First few rows of the dataset:")
print(df.head())

# Show basic info about the dataset
print("\nDataset Info:")
print(df.info())

# Display summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create stacked area chart
fig, ax = plt.subplots(figsize=(12, 7))

# Define colors - bright for electric, muted for others
colors = ["#3F24A1",  # Regular - light gray
          '#B0B0B0',  # Hybrid - medium gray
          '#909090',  # Kids - darker gray
          "#F5D02C",  # Electric - bright orange
          '#A0A0A0']  # Other - gray

# Create stacked area chart
ax.stackplot(df['Year'],
             df['Regular bicycles'],
             df[' Hybrid bicycles'],
             df[' Kids bicycles'],
             df[' Electric bicycles'],
             df[' Other bicycles'],
             labels=['Regular', 'Hybrid', 'Kids', 'Electric', 'Other'],
             colors=colors,
             alpha=0.85)

# Add title and labels
ax.set_title('Electric Bicycles Surge in Dutch Market Share (2013-2023)', 
             fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Market Share', fontsize=12)

# Add annotation for electric growth
ax.annotate('Electric bicycles\ngrew from 19% to 57%',
            xy=(2023, 0.85), xytext=(2020, 0.75),
            fontsize=11, fontweight='bold', color='#FF6B35',
            arrowprops=dict(arrowstyle='->', color='#FF6B35', lw=2),
            bbox=dict(boxstyle='round,pad=0.5', facecolor='white', edgecolor='#FF6B35', lw=2))

# Add legend
ax.legend(loc='upper left', frameon=True, fontsize=10)

# Set y-axis to percentage format
ax.set_ylim(0, 1)
ax.set_yticks([0, 0.25, 0.5, 0.75, 1.0])
ax.set_yticklabels(['0%', '25%', '50%', '75%', '100%'])

# Grid for readability
ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```
