```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Excel file
df_wait = pd.read_excel("Waiting_times.xlsx", sheet_name="Sheet1")

# Create a combined Period column
df_wait["Period"] = df_wait["Month"] + " " + df_wait["Year"].astype(str)

# Plot
plt.figure(figsize=(10,6))

plt.plot(df_wait["Period"], df_wait["Min"], marker="o", label="Min")
plt.plot(df_wait["Period"], df_wait["Avg waiting time"], marker="o", label="Avg")
plt.plot(df_wait["Period"], df_wait["Max"], marker="o", label="Max")

plt.xticks(rotation=45, ha="right")
plt.ylabel("Waiting Time (minutes)")
plt.title("Waiting Times Over Time")
plt.legend()
plt.tight_layout()
plt.show()
```
